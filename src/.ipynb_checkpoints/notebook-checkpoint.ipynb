```python
import nilearn
import nibabel
import pandas as pd
import numpy as np
import torch
```

```python
def hemispheric_data(filepath):

    df = pd.read_csv(
        filepath,
        sep=r'\s+',
        comment='#',
        header=None,
        names=[
            'StructName', 'NumVert', 'SurfArea', 'GrayVol',
            'ThickAvg', 'ThickStd', 'MeanCurv', 'GausCurv',
            'FoldInd', 'CurvInd'
        ]
    )
    return df


def volumetric_data(filepath):
    skip = []
    with open(filepath) as f:
        for i, line in enumerate(f):
            if line.lstrip().startswith('#') or line.strip() == '':
                skip.append(i)
    
    return pd.read_csv(
        filepath,
        sep=r'\s+',
        skiprows=skip,
        header=None,
        names=[
            'Index', 'SegId', 'NVoxels', 'Volume_mm3', 'StructName',
            'normMean', 'normStdDev', 'normMin', 'normMax', 'normRange'
        ]
    )

    
def phenotypic_data(filepath):
    return pd.read_csv(filepath)
```
